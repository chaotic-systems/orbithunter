```python
n = 31
m = 31
L = 88
T= 120

xy = np.meshgrid(np.arange(m),np.arange(n))

spacetime_mollifier = np.exp(-(np.abs(xy[0]-sms)/8 + np.abs(xy[1]-tms)/tms))
spacetime_mollifier = np.concatenate((spacetime_mollifier, spacetime_mollifier), axis=1)
spacetime_mollifier = np.concatenate((spacetime_mollifier, spacetime_mollifier), axis=0)
spacetime_mollifier = np.concatenate((np.ones([1, 2*m]), spacetime_mollifier), axis=0)
```

```python
sms
```

```python
tms = T//10
sms = L // (pi*np.sqrt(2))
sms = int(sms)
tms = int(tms)
mollifier_exponents = sms + -1 * np.tile(np.arange(0, m)+1, (n, 1))
mollifier = np.exp(mollifier_exponents)
mollifier[:, :sms] = 1
mollifier[tms:, :] = 0
mollifier = np.concatenate((mollifier, mollifier), axis=1)
mollifier = np.concatenate((mollifier, mollifier), axis=0)
mollifier = np.concatenate((np.ones([1, 2*m]), mollifier), axis=0)
rmodes = np.random.randn(*spacetime_mollifier.shape)
```
